### Visualize Objective Function

#### 1D Function

```python
#x^2
def objective_function(X):
  return X**2
```

```python
#range_value
import numpy as np
r_min=-5
r_max=5
dif=0.1
X=np.arange(r_min,r_max,dif)
y=objective_function(X)
```

```python
for i in range(5):
  print(f"f({round(X[i],2)})={round(y[i],2)}")
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
plt.show()
```

```python
plt.scatter(X,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
optimal_X=0
optimal_y=objective_function(optimal_X)
plt.scatter(optimal_X,optimal_y,marker='s',color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
optimal_X=0
optimal_y=objective_function(optimal_X)
plt.axvline(optimal_X,optimal_y,color='red',ls='--')
plt.show()
```

```python
np.random.seed(1)
sample=r_min+np.random.rand(10)*(r_max-r_min)
sample_eval=objective_function(sample)
```

```python
plt.plot(X,y)
plt.plot(sample,sample_eval,'o',color='black')
plt.axvline(optimal_X,optimal_y,color='red',ls='--')
```

#### 2D Function

```python
#X^2+Y^2
def objective(x,y):
  return x**2+y**2

X=4
y=4
result=objective(X,y)
print(result)
```

```python
from numpy import arange
from matplotlib import pyplot
from numpy import meshgrid
r_min=-5
r_max=5
diff=0.1
Xaxis=arange(r_min,r_max,diff)
yaxis=arange(r_min,r_max,diff)
X,y=meshgrid(Xaxis,yaxis)
result=objective(X,y)
```

```python
result.shape
```

```python
for i in range(2):
  print(result)
```

```python
plt.contour(X,y,result,50,alpha=1.0,cmap='jet')
```

```python
plt.contourf(X,y,result,50,cmap='jet')
X_optimal=0
y_optimal=0
plt.plot(X_optimal,y_optimal,'*',color='white')
```

```python
X_sample=r_min+np.random.rand(10)*(r_max-r_min)
y_sample=r_min+np.random.rand(10)*(r_max-r_min)
plt.contourf(X,y,result,50,cmap='jet')
X_optimal=0
y_optimal=0
plt.plot(X_optimal,y_optimal,'*',color='white')
plt.plot(X_sample,y_sample,'o',color='black')
```

```python
fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(X,y,result,cmap='jet')
```
